```python
class QuantizedPiecewiseGrid(object):

    def __init__(self, geodf, xstep, ystep):

        self.geodf = geodf
        self.p = so.unary_union(self.geodf.geometry)
        (self.xbins, self.ybins), (self.gxs, self.gys) = gp.overlay_grid(self.p, xstep, ystep)
        
        
    def make_grid(self):
        dxs = np.zeros(self.gxs.shape)
        dys = np.zeros(self.gxs.shape)
        for ii, x in enumerate(self.xbins):
            for jj, y in enumerate(self.ybins):
                pt = Point((x,y))
                vector = self.get_vector(pt)
                dxs[ii, jj] = vector[0]
                dxs[ii, jj] = vector[1]
        return dxs, dys
        
    def get_containing_geom(self, p):
        ind = self.geodf.contains(p)
        ind = np.flatnonzero(ind)
        assert len(ind) == 1, print(ind)
        ind = ind[0]
        return scaled_hatch_pixels.loc[ind, :]
        
    def get_vector(self, p):
        geom_row = self.get_containing_geom(p)
        angle = geom_row['angle']
        angle = np.radians(angle)
        
        intensity = geom_row['intensity']  #not doing anything with this rn
        
        vector = np.array((np.cos(angle), np.sin(angle)))
        return vector
```

```python



# Cell
class VectorParticle(object):

    def __init__(self, pos, grid, vector=None, stepsize=1):
        self.pos = pos
        self.grid = grid
        self.stepsize = stepsize
        self.n_step = 0
        self.pts = [self.pos]
        self.in_bounds = True
        if vector == None:
            vector = np.array([0, 0])
        self.vector = vector

    @property
    def x(self):
        return self.pos.x
    
    @property
    def y(self):
        return self.pos.y
    
    @property
    def xy(self):
        return np.array([self.x, self.y])

    @property
    def line(self):
        return LineString(self.pts)
    
    def update_vector(self):
        self.vector = self.grid.get_vector(self.pos)
        
    def check_if_in_bounds(self):
        self.in_bounds = self.grid.p.contains(self.pos)

    def calc_step(self):
        self.update_vector()
        self.dx = self.vector[0] * self.stepsize
        self.dy = self.vector[1] * self.stepsize


    def step(self):
        self.check_if_in_bounds()
        if self.in_bounds:
            self.calc_step()
            self.pos = sa.translate(self.pos, xoff=self.dx, yoff=self.dy)
            self.pts.append(self.pos)
```

```python
image_path= '/home/naka/art/wigglesphere.jpg'
filename = 'vp_test.svg'
paper_size:str = '11x14 inches'
border:float=15  # mm
image_rescale_factor:float=0.004
smooth_disk_size:int=0
hist_clip_limit=0.1
hist_nbins=32
intensity_min=0.
intensity_max=1.
hatch_spacing_min=0.35  # mm
hatch_spacing_max=1.1 # mm
pixel_width=1 # mm
pixel_height=1 # mm
angle_jitter='ss.norm(loc=10, scale=0).rvs'  # degrees
pixel_rotation='0'  # degrees
merge_tolerances=[0.3, 0.4,]  # mm
simplify_tolerances=[0.2,]  # mm
savedir='/home/naka/art/plotter_svgs'
```

```python
# make page
paper = Paper(paper_size)
drawbox = paper.get_drawbox(border)

# load
img =  rgb2gray(io.imread(Path(image_path)))
```

```python

        
img_rescale = rescale(img, image_rescale_factor)
img_renorm = exposure.equalize_adapthist(img_rescale, clip_limit=hist_clip_limit, nbins=hist_nbins)
# img_renorm = img_rescale
```

```python
# calc dominant angle
selem = disk(smooth_disk_size)
filt_img = filters.rank.mean(img_renorm, selem)
angle_farid = local_angle(filters.farid_h(filt_img), filters.farid_v(filt_img))
```

```python
plt.imshow(angle_farid)
```

```python
# make pixel polys
prms = []
for y, row in tqdm(enumerate(img_renorm)):
    for x, intensity in enumerate(row):

        p = gp.centered_box(Point(x, y), width=pixel_width, height=pixel_height)
        a = np.degrees(angle_farid[y, x])
        prm = {
            'geometry':p,
            'x':x,
            'y':y,
            'raw_pixel_width':pixel_width,
            'raw_pixel_height':pixel_height,
            'intensity': intensity,
            'angle':a,
            'group': 'raw_hatch_pixel',

        }
        prms.append(prm)
raw_hatch_pixels = geopandas.GeoDataFrame(prms)
```

```python
#  rescale polys to fit in drawbox
bbox = box(*raw_hatch_pixels.total_bounds)
_, transform = gp.make_like(bbox, drawbox, return_transform=True)
A = gp.AffineMatrix(**transform)
scaled_hatch_pixels = raw_hatch_pixels.copy()
scaled_hatch_pixels['geometry'] = scaled_hatch_pixels.affine_transform(A.A_flat)
scaled_hatch_pixels['scaled_pixel_height'] = scaled_hatch_pixels['geometry'].apply(gp.get_height)
scaled_hatch_pixels['scaled_pixel_width'] = scaled_hatch_pixels['geometry'].apply(gp.get_width)
```

```python
new_drawbox = so.unary_union(scaled_hatch_pixels.geometry)
```
