```python
# make page
paper_size = '5x7 inches'
border:float=20
paper = Paper('7x5 inches')

drawbox = paper.get_drawbox(border)
```

```python
trcp = gp.RecursiveCirclePacker(drawbox.buffer(20), rad_seq_start=0.12,
    rad_seq_end=0.12, min_allowed_rad=2, n_rads=10, )
trcp.run(1)
```

```python
vertices = MultiPoint([c.centroid for c in trcp.unfilled_circles])
```

```python
tris = so.triangulate(vertices)
```

```python
def poly_to_edges(poly):
    bcoords = poly.boundary.coords
    n_edges = len(bcoords)-1
    edges = [LineString((bcoords[i], bcoords[i+1])) for i in range(n_edges)]
    return edges
```

```python
def inscribe_poly(edges, interp_distances):
    pts = [edge.interpolate(d, normalized=True) for edge,d in zip(edges, interp_distances)]
    return Polygon(pts)
```

```python
f = gp.make_callable(0.5)
```

```python
new_polys = []
for poly in tris:
    edges = poly_to_edges(poly)
    interp_distances = [0.1, 0.5, 0.9]
    inscribed_poly = inscribe_poly(edges, interp_distances)
    subdivided_poly = gp.merge_Polygons([poly.difference(inscribed_poly), inscribed_poly]).buffer(-0.01)
    new_polys.append(subdivided_poly)
    
polys = [p for p in gp.merge_Polygons(new_polys)]
```

```python
new_polys = []
for i,poly in enumerate(polys):
    edges = poly_to_edges(poly)
    interp_distances = [0.1, 0.5, 0.9]
    inscribed_poly = inscribe_poly(edges, interp_distances)
    subdivided_poly = [poly.difference(inscribed_poly)] + [inscribed_poly]
    new_polys += subdivided_poly
    
polys = gp.merge_Polygons(new_polys)
```
